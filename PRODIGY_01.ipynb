# **1) IMPORTING PACKAGES**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# **2) LOADING DATA AND INITIAL INSPECTION**

```python
# Load the data
data = pd.read_csv('/content/population - female.csv',skiprows=4)
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Display the shape of the dataset (number of rows and columns)
print(f"Shape of the dataset: {data.shape}\n")
```

```python
# Get information about the dataset (data types, non-null values, etc.)
print(data.info())
```

```python
# Display summary statistics for the dataset
print(data.describe())
```

# **3) DATA CLEANING/PREPROCESSING**

```python
# Drop null columns
data = data.drop(columns=['Unnamed: 68'])
```

```python
# Remove rows with null values
data_cleaned = data.dropna()

# Shape of the data after cleaning
print(f"Shape of the data after cleaning: {data_cleaned.shape}")

# Display the first few rows of the cleaned data
print(data_cleaned.head())
```

# **4) DATA VISUALISATION**

```python
# Select relevant columns (Country Name, Year, and Population)
data_2023 = data_cleaned[['Country Name','Country Code' , '2023']]  # Using the most recent year available, e.g., 2023
```

```python
# Create a bar chart
plt.figure(figsize=(14, 8))
plt.bar(data_2023['Country Name'], data_2023['2023'], color='skyblue')

# Add labels and title
plt.xlabel('Country Name')
plt.ylabel('Female Population')
plt.title('Female Population by Country in 2023')
plt.xticks(rotation=90)  # Rotate country names for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Sort the data by female population in descending order
data_sorted = data_2023.sort_values(by='2023', ascending=False)

# Select the top 20 countries
top_20 = data_sorted.head(20)

# Create a bar chart for the top 20 countries
plt.figure(figsize=(14, 8))
plt.bar(top_20['Country Name'], top_20['2023'], color='skyblue')

# Add labels and title
plt.xlabel('Country Name')
plt.ylabel('Female Population')
plt.title('Top 20 Countries by Female Population in 2023')
plt.xticks(rotation=90)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Create an interactive bar chart
fig = px.bar(data_2023, x='Country Name', y='2023', title='Female Population by Country in 2023')
fig.update_layout(xaxis_title='Country Name', yaxis_title='Female Population')
fig.show()
```

```python
# Filter data for India
india_data = data_cleaned[data_cleaned['Country Name'] == 'India']

# Remove unnecessary columns and melt the data frame
india_data = india_data.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
india_data_melted = india_data.melt(id_vars=['Country Name'], var_name='Year', value_name='Female Population')

# Sort the data by Year
india_data_melted = india_data_melted.sort_values(by='Year')

# Create a line plot for the change in female population over time
plt.figure(figsize=(14, 8))
plt.plot(india_data_melted['Year'], india_data_melted['Female Population'], marker='o', linestyle='-', color='skyblue')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Female Population')
plt.title('Change in Female Population in India (1960-2023)')
plt.xticks(rotation=90)
plt.grid(True)
```

```python
# Create a bar chart for the top 20 countries
plt.figure(figsize=(14, 8))
plt.bar(india_data_melted['Year'], india_data_melted['Female Population'], color='skyblue')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Female Population')
plt.title('Change in Female Population in India (1960-2023)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
